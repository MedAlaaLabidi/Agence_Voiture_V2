```python
import pandas as pd
```

```python
df=pd.read_excel("dataset.xlsx",index_col=0)
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe=OneHotEncoder()
ohe.fit(df[['marque']])

```

```python
ohe1=OneHotEncoder()
ohe1=ohe1.fit(df[['couleur']])
```

```python
ohe1.categories_[0]
```

```python
ohe.categories_[0]
```

```python
couleur_sparse=ohe1.transform(df[['couleur']])
couleurs=couleur_sparse.toarray()
couleurs
```

```python
marques_sparse=ohe.transform(df[['marque']])
marques=marques_sparse.toarray()
marques
```

```python
df_marque=pd.DataFrame(marques,index=df.index,columns='marque_'+ohe.categories_[0])
df_marque
```

```python
df_couleur=pd.DataFrame(couleurs,index=df.index,columns='couleur_'+ohe1.categories_[0])
df_couleur
```

```python
df=pd.concat([df,df_marque,df_couleur], axis=1)
df
```

```python
df.drop('marque', axis=1, errors='ignore', inplace=True)
df.drop('couleur', axis=1, errors='ignore', inplace=True)

df
```
